```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import xlwings as xw
```

```python
def goodinfo_crawler(stock_id):    
    url = f"https://goodinfo.tw/StockInfo/ShowK_Chart.asp?STOCK_ID={stock_id}&CHT_CAT2=DATE"
    header = {
    "user-agent": "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.106 Mobile Safari/537.36",
    }
    res = requests.get(url, headers = header)
    #print(res.encoding) 印出request使用何種解碼
    #print(res.apparent_encoding) 印出網站如何編碼
    res.encoding = "utf-8"
    soup = BeautifulSoup(res.text, "html.parser")
    date_list = list()
    price_list = list()
    for i in range(54):
        table_row = soup.find_all("tr", {"id":f"row{i}"})[0]
        date = table_row.find_all("td")[0].text
        close_price = table_row.find_all("td")[4].text
        date_list.append(date)
        price_list.append(close_price)
    date_list.reverse()
    price_list.reverse()
    return date_list, price_list
```

```python
df = pd.DataFrame(goodinfo_crawler(2337)[1], index = goodinfo_crawler(2337)[0], columns = ["close_price"])
df
```

```python
df["five_day_ma"] = df["close_price"].rolling(5).mean()
df
```
